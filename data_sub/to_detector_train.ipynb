```python
import pandas as pd
import numpy as np
import json
import seaborn as sns
import os
import cv2
from matplotlib import pyplot as plt
```

```python
#Detection Result Json
data_json_train = '/opt/ml/detection/dataset/train_fold1.json'
```

```python
with open(data_json_train, 'r') as outfile:
    data = (json.load(outfile))

#images
data_image = data['images']
data_anno = data['annotations']
data_categories = data['categories']
data_image_df = pd.DataFrame(data_image)
data_anno_df = pd.DataFrame(data_anno)
data_categories_df = pd.DataFrame(data_categories)
data_anno_count = data_anno_df['image_id'].value_counts().reset_index()
data_anno_count.columns = ['id', 'counts']

data_image_df = data_image_df.merge(data_anno_count, how='left', on='id')
data_image_under_3 = data_image_df.loc[data_image_df['counts'] <= 3]
```

```python
#anotations
data_id = data_image_under_3['id'].to_frame()
data_id.columns=['image_id']
data_anno_under_3 = data_id.merge(data_anno_df, how='left', on='image_id')
```

```python
data_categories
```

```python
d_image_json = data_image_under_3.to_json(orient='records')
d_annotation_json = data_anno_under_3.to_json(orient='records')
d_categories_json = data_categories_df.to_json(orient='records')

parsed_img = json.loads(d_image_json)
parsed_anno = json.loads(d_annotation_json)
parsed_categories = json.loads(d_categories_json)

final_json = json.dumps({'images':parsed_img, 'annotations':parsed_anno, 'categories':parsed_categories}, indent=4)

parsed = json.loads(final_json)

file_path = '/opt/ml/detection/dataset/extract_json/train_under3.json' #경로 매번 수정
with open(file_path, 'w') as outfile:
    json.dump(parsed, outfile, indent=4)
parsed
```
